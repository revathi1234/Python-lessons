### Array values from user

```python
from array import *   #import array 

arr = array('i',[])  #create an empty array first
n = int(input('enter the length of array'))  #ask user to enter length of array 

for i in range(n):
    x = int(input('enter the next value'))
    arr.append(x)
    
    
print(arr)


```

```python
##### to get index number 

 
val=int(input('enter value to search'))

k=0
for e in arr:
    if e==val:
        print('index is',k)
        break
        
    k=k+1
        
```

```python
 #using function
print(arr.index(val))
```
