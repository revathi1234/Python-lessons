## Pattern printing

```python
for i in range(4):
    for j in range(4):
        print("#",end=" ")
        
    print()
    
   
```

```python
for i in range(4):
    for j in range(i+1):
        print("#",end=" ")
        
    print()
    
   
```

#### Uisng sqrt

```python
n=int(input("Enter a number to check : "))
for i in range(2,int(n**0.5)+1):
    if n%i==0:
        print("Not Prime")
        break
else:
    print("Prime")
```

By checking for factors only up to the square root of  n, we cover all possible factor pairs without redundant checking. 
If n is not divisible by any number up to its square root, then it won't be divisible by any larger number either. 
This is due to the fact that any factors larger than the square root would have corresponding factors smaller than the square root.

## Fibonnaci series

0 1 1 2 3 5 8 13

```python

def fib(n):
    
    a=0
    b=1
    
    if n<0:
        print('invalid number')
    elif n==1:
        print(a)
        
    else:
        print(a)
        print(b)
    
        for i in range(2,n):
            c=a+b
            a=b
            b=c
            print(c)
        
    
   

fib(30)
```

```python
def fact(n):
    
    if(n==0):
        return 1
    else:
        return n*fact(n-1)
    

result=fact(71)

print(result)
```
